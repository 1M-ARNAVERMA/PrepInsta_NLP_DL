```python
import pandas as pd
import numpy as np
import matplotlib as plt
import tensorflow as tf
```

```python
tf.__version__
```

```python
dataset = pd.read_csv('American_Express_User_Exit_Pred.csv')
x = dataset.iloc[:,0:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

## Gender Column - Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
x[:,2] = label_encoder.fit_transform(x[:,2])
```

```python
print(x)
```

## Geographical Column - One hot encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[1])],remainder = 'passthrough')
x = np.array(ct.fit_transform(x))
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.20 , random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

## ANN

```python
# INITIALIZATION
ann = tf.keras.models.Sequential()
```

```python
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(128, activation='relu', input_shape=(x_train.shape[1],)))
ann.add(tf.keras.layers.Dense(64, activation='relu'))
ann.add(tf.keras.layers.Dense(5, activation='softmax'))
```

### ANN Training

```python
ann.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
ann.fit(x_train, y_train, batch_size=32, epochs=120)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
print(x_train.shape)
```

## Single Prediction 

```python
y_pred = ann.predict(x_test)
y_pred = np.argmax(y_pred, axis=1)

print(
    np.concatenate(
        (y_pred.reshape(-1, 1), y_test.reshape(-1, 1)),
        axis=1
    )
)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
